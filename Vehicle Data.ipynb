# Vehicle Data Preparation

## source
- https://www.fueleconomy.gov/feg/download.shtml

```python
# Importing library
import pandas as pd
```

```python
# Importing data source
car_infodf = pd.read_csv('car_info.csv')
```

```python
car_infodf.head(5)
```

```python
car_infodf["Brand"] = car_infodf["Model"].str.split(" ").str[0]
car_infodf['Model'] = car_infodf['Model'].str.split(n=1).str[1]
car_infodf.head(5)
```

```python
comb_car_df = car_infodf[['Brand', 'Model', 'Drive', 'Fuel', 'Stnd', 'City MPG', 'Hwy MPG']]
comb_car_df
```

```python
comb_car_df.to_csv('comb_car_info.csv')
```
